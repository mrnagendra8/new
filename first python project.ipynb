```python


# Count the number of calls with call_status "Answered"
connected_calls = df[df["Call Status"] == "Answered"].shape[0]

print("Number of connected calls:", connected_calls)
```

```python
# Filter the DataFrame to include only rows where call_status is "Answered"
answered_calls_df = df[df['Call Status'] == "Answered"]

# Count the number of unique lead IDs
unique_connected_calls = answered_calls_df['Lead Id'].nunique()

print("Unique calls connected:", unique_connected_calls)
```

```python
# Count the number of calls with call_status "Interested"
converted_leads = df[df['Call Status'] == "Interested"].shape[0]

print("Number of converted leads:", converted_leads)
```

```python
# Define criteria for qualified leads
criteria = (df['Call Status'] == "Interested") & (df['demographic_criteria'] == "met")

# Filter the DataFrame based on criteria
qualified_leads_df = df[criteria]

# Count the number of qualified leads
qualified_leads_count = qualified_leads_df.shape[0]

print("Qualified leads:", qualified_leads_count)
```

```python
# Count the number of calls with call_status "Not Interested"
lost_leads = df[df['Call Status'] == "Not Interested"].shape[0]

print("Number of lost leads:", lost_leads)
```

```python
#Calculate the average Agent Call Duration
average_agent_call_duration = df['Agent Call Duration'].mean()

print("Average agent call duration:", average_agent_call_duration)
```

```python
# Calculate the average customer call duration
average_customer_call_duration = df['Customer Call Duration'].mean()

print("Average customer call duration:", average_customer_call_duration)
```
